```python
# Importar librerias
from keras.preprocessing import image
from keras import models
import numpy as np
import matplotlib.pyplot as plt

# Importar el modelo
model = models.load_model("flowersVGG.h5")
```

```python
# Imagenes de prueba
#test_img = '/daisy/daisy.jpg'
#test_img = '/dandelion/dandelion.jpg'
#test_img = '/sunflower/image14.jpeg'
test_img = '/tulip/tulipanT.jpg'

# Etiquetas
label = ['daisy','dandelion','sunflower','tulip']

img_path = 'Flowers/test'+ test_img

# Procesamiento de la imagen
img = image.load_img(img_path,  target_size=(150,150))
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis = 0)
img_tensor /= 255.

# Respuesta
confidence = model.predict(img_tensor)
predict_class = np.argmax(confidence)
print (confidence)
print ("class ", label[predict_class], predict_class+1)

plt.imshow(img_tensor[0])
plt.show()
```
